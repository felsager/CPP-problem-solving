```python
import matplotlib.pyplot as plt
```

```python
# Data points
x_values = [1000, 2000, 3500, 3600, 4000, 5000, 10000, 20000, 50000]
y_values = [0.014, 0.054, 0.164, 0.173, 0.215, 0.334, 1.327, 8.93, 92.6]


# Plotting the data with a log scale on the y-axis
plt.scatter(x_values, y_values, label='Data Points')
plt.yscale('log')  # Set log scale on the y-axis

# Adding labels and title
plt.xlabel('X Values')
plt.ylabel('Y Values (log scale)')
plt.title('Scatter Plot of Data Points with Logarithmic y-axis')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
